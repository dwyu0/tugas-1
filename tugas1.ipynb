1. Inputkan 10 nilai

```python
arr = []

for i in range(10):
    nilai = int(input("Masukkan nilai: "))
    arr.append(nilai)
```

2. Tampilkan nilai tertinggi & terendah

```python
print("Nilai maksimum:", max(arr))
print("Nilai minimum:", min(arr))

```

3. Hitung rata-rata

```python
print("Nilai rata-rata:",sum(arr)/len(arr))
```

4. Hitung jumlah mahasiswa lulus (>= 60)

```python
lulus = sum(1 for n in arr if n >= 60)
tidak_lulus = len(arr) - lulus

#OUTPUT
print("\n=== HASIL ===")
print("Nilai:", arr)
print("Nilai tertinggi:", max(arr))
print("Nilai terendah:", min(arr))
print("Rata-rata:", sum(arr)/len(arr))
print("Jumlah lulus:", lulus)
print("Jumlah tidak lulus:", tidak_lulus)
```

5. Buat grafik untuk menggambarkan nilai tertinggi dan terendah

```python
import matplotlib.pyplot as plt
plt.figure()
plt.bar(["Tertinggi", "Terendah"], [max(arr), min(arr)])
plt.title("Grafik Nilai Tertinggi & Terendah")
plt.ylabel("Nilai")
```

6. Buat grafik untuk menggambarkan data kelulusan

```python
plt.figure()
plt.pie([lulus, tidak_lulus], labels=["Lulus", "Tidak Lulus"], autopct='%1.1f%%')
plt.title("Grafik Kelulusan")

plt.show()
```
